## 3. Data Wrangling

```python
# Identify which columns are useful to extract business insights
# (not technical, not IDs, not noise)
fed_df= df[
[
   'country',
    'managed_by',
    'fulfill_via',
    'shipment_mode',
    'pq_first_sent_to_client_date',
    'po_sent_to_vendor_date',
    'scheduled_delivery_date',
    'delivered_to_client_date',
    'delivery_recorded_date',
    'line_item_quantity',
    'line_item_value',
    'weight_kilograms',
    'freight_cost_usd',
    'vendor',
    'line_item_insurance_usd'
]
]
```

```python
fed_df.head()
```

## Cleaning Datatypes

```python
fed_df.dtypes
```

```python
# Date datatype varaibles 
date_cols = [
    'pq_first_sent_to_client_date',
    'po_sent_to_vendor_date',
    'scheduled_delivery_date',
    'delivered_to_client_date',
    'delivery_recorded_date']

for col in date_cols:
    fed_df[col] = pd.to_datetime(fed_df[col], errors='coerce')

# Numerical Columns
num_cols = [
    'line_item_quantity',
    'line_item_value',
    'weight_kilograms',
    'freight_cost_usd',
    'line_item_insurance_usd']

for col in num_cols:
    fed_df[col] = pd.to_numeric(fed_df[col], errors='coerce')

# Catergorical variables
cat_cols = [
    'country',
    'managed_by',
    'fulfill_via',
    'shipment_mode',
    'vendor']

for col in cat_cols:
    fed_df[col] = (
        fed_df[col]
        .astype(str)
        .str.strip()
        .str.title())
```

```python
fed_df.dtypes
```

```python
date_cols=['pq_first_sent_to_client_date','po_sent_to_vendor_date','scheduled_delivery_date',
           'delivered_to_client_date','delivery_recorded_date']
for col in date_cols:fed_df[col]=pd.to_datetime(fed_df[col],errors='coerce')

num_cols=['line_item_quantity','line_item_value','weight_kilograms','freight_cost_usd',
          'line_item_insurance_usd']
for col in num_cols:fed_df[col]=pd.to_numeric(fed_df[col],errors='coerce')

cat_cols=['country','managed_by','fulfill_via','shipment_mode','vendor']
for col in cat_cols:fed_df[col]=fed_df[col].astype(str).str.strip().str.title()
```

##  4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

## Univariate Analysis

```python
# BASIC STYLING
sns.set(style="whitegrid")
palette = sns.color_palette("Set2")
# CHART: COUNT OF SHIPMENTS BY COUNTRY
plt.figure(figsize=(6,6))

sns.countplot(
    x='country',
    data=fed_df,
    order=fed_df['country'].value_counts().nlargest(10).index,
    palette=palette)
# TITLES & LABELS
plt.title('Count of Shipment by Country', fontsize=18, fontweight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of Shipments', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Distribution of Shipment Modes

```python
# Count shipment modes
shipment_mode_counts = fed_df['shipment_mode'].value_counts()

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Explode second slice slightly (for emphasis)
explode = [0.1 if i == 1 else 0 for i in range(len(shipment_mode_counts))]

# Plot pie chart
ax.pie(
    shipment_mode_counts,
    labels=shipment_mode_counts.index,
    colors=palette,
    explode=explode,
    autopct='%.0f%%',
    startangle=90
)

# Title
ax.set_title(
    'Shipment Mode Distribution',
    fontsize=18,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```

## Average Freight Cost

```python
# Calculate average freight cost
avg_freight_cost = fed_df['freight_cost_usd'].mean()

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Histogram with KDE
sns.histplot(
    x='freight_cost_usd',
    data=fed_df,
    kde=True,
    bins=30,
    color='#4d148c',
    ax=ax)

# Titles and labels
ax.set_title(
    'Freight Cost Distribution',
    fontsize=16,
    fontweight='bold'
)
ax.set_xlabel('Freight Cost (USD)')
ax.set_ylabel('Frequency')

# Add average line
ax.axvline(
    avg_freight_cost,
    color='red',
    linestyle='--',
    linewidth=2,
    label=f'Average = ${avg_freight_cost:,.0f}'
)

ax.legend()

plt.tight_layout()
plt.show()
```

## Line Item Quantity Distribution

```python
# Count frequency of each line item quantity
count_line_quantity = fed_df['line_item_quantity'].value_counts()

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Histogram with KDE
sns.histplot(
    x='line_item_quantity',
    data=fed_df,
    kde=True,
    bins=30,
    color='#ff6600',
    ax=ax
)

# Titles and labels
ax.set_title(
    'Line Item Quantity Distribution',
    fontsize=16,
    fontweight='bold'
)
ax.set_xlabel('Line Item Quantity')
ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Weight Distribution (Kilograms)

```python
# ================================
# Chart - 5: Shipment Weight Distribution
# ================================

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Boxplot of shipment weights
sns.boxplot(
    x='weight_kilograms',
    data=fed_df,
    color='#4d148c',
    showfliers=False,
    ax=ax
)

# Titles and labels
ax.set_title(
    'Shipment Weight Distribution',
    fontsize=16,
    fontweight='bold'
)
ax.set_xlabel('Weight (Kilograms)')
ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

## Vendors with highest orders

```python
# ================================
# Chart - 6: Vendors with Highest Orders
# ================================

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

sns.countplot(
    x='vendor',
    data=fed_df,
    order=fed_df['vendor'].value_counts().nlargest(5).index,
    palette=palette,
    ax=ax
)

# Titles and labels
ax.set_title(
    'Vendors with Highest Orders',
    fontsize=16,
    fontweight='bold'
)
ax.set_xlabel('Vendor')
ax.set_ylabel('Number of Orders')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

# Bi-Variate Analysis

## Average Weight by Shipment Mode

```python
# Chart - 11: Average Weight by Shipment Mode

fig, ax = plt.subplots(figsize=(10, 6))

sns.barplot(
    data=fed_df,
    x='shipment_mode',
    y='weight_kilograms',
    estimator='mean',
    palette=palette,
    ax=ax
)

ax.set_title('Average Weight by Shipment Mode', fontsize=16, fontweight='bold')
ax.set_xlabel('Shipment Mode')
ax.set_ylabel('Average Weight (Kilograms)')
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right')

plt.tight_layout()
plt.show()
```

## Delivery Delays by Shipment Mode

```python
# Chart - 7 visualization code
fed_df['delay_days'] = (fed_df['delivered_to_client_date'] - fed_df['scheduled_delivery_date']).dt.days
fig,ax = plt.subplots(figsize=(10,6))
sns.barplot(x='shipment_mode',y='delay_days',data=fed_df,palette=palette,ax=ax)
ax.set_title('Delivery Delays by Shipment Mode',fontsize=16,fontweight='bold')
ax.set_xlabel('Shipment Mode')
ax.set_ylabel('Average Delay (Days)')
ax.set_xticklabels(ax.get_xticklabels(),rotation=30,ha='right')
plt.tight_layout()
plt.show()
```

## Total Shipments Over Time

```python
# Chart - Total Shipments Over Time

fed_df['shipment_date'] = fed_df['pq_first_sent_to_client_date'].fillna(fed_df['scheduled_delivery_date'])
shipments_over_time = fed_df.resample('M', on='shipment_date').size()

fig, ax = plt.subplots(figsize=(10,6))

ax.fill_between(
    shipments_over_time.index,
    shipments_over_time.values,
    color='#ff6600',
    alpha=0.6
)

ax.set_title('Total Shipments Over Time',fontsize=16,fontweight='bold')
ax.set_xlabel('Time')
ax.set_ylabel('Number of Shipments')

plt.tight_layout()
plt.show()
```

## Line Of Items Across Diffrent Shipment Mode

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.boxplot(
    x='shipment_mode',
    y='line_item_value',
    data=fed_df,
    palette=palette,
    showfliers=False,
    ax=ax
)

ax.set_title('Line Item Value Across Different Shipment Modes',fontsize=16,fontweight='bold')
ax.set_xlabel('Shipment Mode')
ax.set_ylabel('Line Item Value')
ax.set_xticklabels(ax.get_xticklabels(),rotation=30,ha='right')

plt.tight_layout()
plt.show()
```

## Freight Cost In Comparison to Weight

```python
# Create figure
fig,ax=plt.subplots(figsize=(10,6))
# Scatter plot
sns.scatterplot(
    x='weight_kilograms',
    y='freight_cost_usd',
    data=fed_df,
    s=250,
    alpha=0.7,
    color='#4d148c',
    ax=ax)
# Titles and labels
ax.set_title('Freight Cost vs Shipment Weight',fontsize=16,fontweight='bold')
ax.set_xlabel('Weight (Kilograms)')
ax.set_ylabel('Freight Cost (USD)')
# Layout adjustment
plt.tight_layout()
plt.show()
```

## Highest Paying Countries based on Freight Cost

```python
# Calculate total freight cost per country

fed_df['freight_cost_usd'] = fed_df['freight_cost_usd'].fillna(0)
fed_df['freight_cost_usd'] = np.where(np.isfinite(fed_df['freight_cost_usd']),fed_df['freight_cost_usd'],0)
fed_df['freight_cost_usd'] = fed_df['freight_cost_usd'].astype(int)

top_10_countries=fed_df.groupby('country')['freight_cost_usd'].sum().nlargest(10).reset_index()
fig,ax=plt.subplots(figsize=(10,6))
sns.barplot(x='country',y='freight_cost_usd',data=top_10_countries,palette=palette,ax=ax)
ax.set_title('Highest Paying Countries Based on Freight Cost',fontsize=16,fontweight='bold')
ax.set_xlabel('Country')
ax.set_ylabel('Total Freight Cost (USD)')
ax.set_xticklabels(ax.get_xticklabels(),rotation=30,ha='right')
plt.tight_layout()
plt.show()
```

## Shipment Mode in comparison to Freight Cost

```python
# Create figure
fig,ax=plt.subplots(figsize=(10,6))

# Box plot
sns.boxplot(
    x='shipment_mode',
    y='freight_cost_usd',
    data=fed_df,
    palette=palette,
    showfliers=False,
    ax=ax
)
# Titles and labels
ax.set_title('Freight Cost by Shipment Mode',fontsize=16,fontweight='bold')
ax.set_xlabel('Shipment Mode')
ax.set_ylabel('Freight Cost (USD)')

# Rotate x-axis labels
ax.set_xticklabels(ax.get_xticklabels(),rotation=30,ha='right')

# Layout adjustment
plt.tight_layout()
plt.show()
```

# Multi-Variate Analysis

## Freight Cost by Country, Shipment Mode, and Vendor

```python
# Select top 10 countries by shipment count
top_10_countries=fed_df['country'].value_counts().nlargest(10).index
top_10_data=fed_df[fed_df['country'].isin(top_10_countries)]

# Create figure
fig,ax=plt.subplots(figsize=(10,6))

# Scatter plot
sns.scatterplot(
    x='weight_kilograms',
    y='freight_cost_usd',
    hue='country',
    size='weight_kilograms',
    style='shipment_mode',
    sizes=(250,1000),
    data=top_10_data,
    palette=palette,
    ax=ax,
    legend=False
)

# Titles and labels
ax.set_title('Freight Cost vs Weight by Shipment Mode and Top 10 Countries',
             fontsize=16,fontweight='bold')
ax.set_xlabel('Weight (Kilograms)')
ax.set_ylabel('Freight Cost (USD)')

# Layout adjustment
plt.tight_layout()
plt.show()

```

## Line Item Value in comaprison with Freight Cost by Vendor and Shipment Mode

```python
# Create figure
fig,ax=plt.subplots(figsize=(10,6))

# Scatter plot by vendor and shipment mode
sns.scatterplot(
    x='line_item_value',
    y='freight_cost_usd',
    hue='vendor',
    style='shipment_mode',
    data=fed_df,
    palette=palette,
    legend=False,
    ax=ax)
# Regression trend line
sns.regplot(
    x='line_item_value',
    y='freight_cost_usd',
    data=fed_df,
    scatter=False,
    color='gray',
    ax=ax)
# Titles and labels
ax.set_title(
    'Line Item Value vs Freight Cost by Vendor and Shipment Mode',
    fontsize=16,
    fontweight='bold'
)
ax.set_xlabel('Line Item Value')
ax.set_ylabel('Freight Cost (USD)')

# Layout adjustment
plt.tight_layout()
plt.show()
```
